```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
from fake_useragent import UserAgent
import time
```

### Let's do web scrapping of cian.ru: we need to collect info about renting a one-room apartment in Moscow

#### Firstly, we'll collect the apartment links by going through all the pages with Selenium

```python
url = "https://www.cian.ru/snyat-1-komnatnuyu-kvartiru/"

options = webdriver.ChromeOptions()
#options.add_argument('--headless')  # Запуск в headless режиме (без GUI)
driver = webdriver.Chrome(options=options)

driver.get(url)
time.sleep(5)

soup = BeautifulSoup(driver.page_source, "html.parser")
links = []
for a in soup.find_all("a", class_="_93444fe79c--link--eoxce", target="_blank"):
    if "href" in a.attrs:
        links.append(a["href"])

for link in links:
    print(link)

driver.quit()
```

```python
#к предыдущей ячейке пробуем добавить переходы по страничкам
url = "https://www.cian.ru/snyat-1-komnatnuyu-kvartiru/"

options = webdriver.ChromeOptions()
#options.add_argument('--headless')  # Запуск в headless режиме (без GUI)
driver = webdriver.Chrome(options=options)

driver.get(url)
time.sleep(5)

all_links = []

while True:
    soup = BeautifulSoup(driver.page_source, "html.parser")
    links = []
    for a in soup.find_all("a", class_="_93444fe79c--link--eoxce", target="_blank"):
        if "href" in a.attrs:
            links.append(a["href"])
            all_links.append(a["href"])
    try:
        next_button = driver.find_element("xpath", "//a[.//span[text()='Дальше']]")
        next_button.click()
        time.sleep(5)
    except Exception as e:
        print("Больше страниц нет:", e)
        break

driver.quit()
len(all_links)
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)

for url in all_links:
    print(url)
    driver.get(url)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    try:
        tube_station = soup.find('a', class_='a10a3f92e9--underground_link--VnUVj').text
        print(tube_station)
    except: 
        print('Error')
        
    time.sleep(1)
```
