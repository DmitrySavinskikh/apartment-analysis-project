```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

### Let's do web scrapping of cian.ru: we need to collect info about renting a one-room apartment in Moscow

#### Firstly, we'll collect the apartment links by going through all the pages with Selenium

```python
url = "https://www.cian.ru/snyat-1-komnatnuyu-kvartiru/"

options = webdriver.ChromeOptions()
#options.add_argument('--headless')  # Запуск в headless режиме (без GUI)
driver = webdriver.Chrome(options=options)

driver.get(url)
time.sleep(5)

soup = BeautifulSoup(driver.page_source, "html.parser")
links = []
for a in soup.find_all("a", class_="_93444fe79c--link--eoxce", target="_blank"):
    if "href" in a.attrs:
        links.append(a["href"])

for link in links:
    print(link)

driver.quit()
```
